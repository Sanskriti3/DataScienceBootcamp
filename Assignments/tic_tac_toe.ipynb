```python
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board, player):
    for row in board:
        if all(cell == player for cell in row):
            return True

    for col in range(3):
        if all(board[row][col] == player for row in range(3)):
            return True

    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

def is_board_full(board):
    return all(all(cell != ' ' for cell in row) for row in board)

def computer_move(board, computer_symbol):
    empty_cells = [(row, col) for row in range(3) for col in range(3) if board[row][col] == ' ']
    return random.choice(empty_cells)

def play_game():
    print("Welcome to Tic-Tac-Toe!")
    player_name = input("Enter your name: ")
    player_symbol = input("Choose X or O: ").upper()
    computer_symbol = 'X' if player_symbol == 'O' else 'O'

    players = {
        'player': {'name': player_name, 'symbol': player_symbol},
        'computer': {'name': 'Computer', 'symbol': computer_symbol}
    }

    turn = random.choice(['player', 'computer'])
    print(f"{players[turn]['name']} goes first!")

    board = [[' ' for _ in range(3)] for _ in range(3)]

    while True:
        print_board(board)

        if turn == 'player':
            row = int(input(f"{players[turn]['name']}'s turn. Enter row (0, 1, or 2): "))
            col = int(input(f"Enter column (0, 1, or 2): "))
        else:
            print(f"{players[turn]['name']}'s turn (Computer)")
            row, col = computer_move(board, players[turn]['symbol'])
            print(f"Computer chose row {row} and column {col}")

        if board[row][col] != ' ':
            print("Cell already filled. Try again.")
            continue

        board[row][col] = players[turn]['symbol']

        if check_winner(board, players[turn]['symbol']):
            print_board(board)
            print(f"{players[turn]['name']} wins!")
            break

        if is_board_full(board):
            print_board(board)
            print("It's a draw!")
            break

        turn = 'computer' if turn == 'player' else 'player'

    play_again = input("Do you want to play again? (yes/no): ")
    if play_again.lower() == 'yes':
        play_game()
    else:
        print("Thanks for playing!")

play_game()
```
